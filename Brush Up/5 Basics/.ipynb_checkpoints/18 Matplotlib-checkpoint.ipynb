```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 6, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

```python
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, 'o--r')
plt.show()
```

```python
plt.plot(ypoints, 'o:m')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
# For 2 lines
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
#Sub Plots

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
# Scatter plots with heatmap

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

```python
# Piecharts

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y,startangle = 90, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, explode=myexplode, labels=mylabels, pctdistance=100, labeldistance=1.2)
plt.legend()
plt.show() 
```
